```python
import numpy as np
student_names=[]
student_scores=[]
def input_score():
    global student_names,student_scores
    student_names.clear()
    student_scores.clear()
    try:
        count=int(input("请输入学生人数："))
        if count<=0:
            print("输入错误")
            return
        for i in range(count):
            name=input(f"\n请输入第{i+1}个学生姓名：")
            score=float(input("请输入成绩："))
            if 0<=score<=100:
                student_names.append(name)
                student_scores.append(score)
            else:
                print("成绩输入错误，该条作废")
        print(f"\n成功录入{len(student_names)}名学生成绩")
    except ValueError:
        print("输入格式错误，需要输入数字")

def show_statistics():
    if not student_scores:
        print("暂无成绩")
        return
    scores_np=np.array(student_scores)
    print(f"""成绩统计
    总人数：{len(scores_np)}
    平均分：{np.mean(scores_np):.2f}
    最高分：{np.max(scores_np):.2f}
    最低分：{np.min(scores_np):.2f}
    中位数：{np.median(scores_np):.2f}
    方差：{np.var(scores_np):.2f}
    标准差：{np.std(scores_np):.2f}
    """)

def show_rank():
    if not student_scores:
        print("暂无成绩数据，请先录入！")
        return
    stu_list = list(zip(student_names, student_scores))
    stu_list.sort(key=lambda x: x[1], reverse=True)
    print("\n===== 成绩排名（从高到低） =====")
    print(f"{'排名':<6}{'姓名':<8}{'成绩':<6}")
    for idx, (name, score) in enumerate(stu_list, start=1):
        print(f"{idx:<6}{name:<8}{score:<6.1f}")

def score_level(score):
    if score >= 90:
        return "优秀"
    elif score >= 60:
        return "及格"
    else:
        return "不及格"
    
def show_score_distribution():
    """功能4：成绩等级分布统计"""
    if not student_scores:
        print("暂无成绩数据，请先录入！")
        return
    excellent = pass_ = fail = 0
    for s in student_scores:
        level =score_level(s)
        if level == "优秀":
            excellent += 1
        elif level == "及格":
            pass_ += 1
        else:
            fail += 1
    total = len(student_scores)
    print("\n===== 成绩等级分布 =====")
    print(f"优秀(90~100)：{excellent}人，占比{excellent/total*100:.1f}%")
    print(f"及格(60~89)：{pass_}人，占比{pass_/total*100:.1f}%")
    print(f"不及格(<60)：{fail}人，占比{fail/total*100:.1f}")

def query_student():
    """功能5：根据姓名查询学生成绩与等级"""
    if not student_scores:
        print("暂无成绩数据，请先录入！")
        return
    target_name = input("请输入要查询的学生姓名：")
    if target_name not in student_names:
        print(f"未找到学生【{target_name}】！")
        return
    idx = student_names.index(target_name)
    score = student_scores[idx]
    level = score_level(score)
    print(f"\n学生：{target_name}")
    print(f"成绩：{score:.1f}，等级：{level}")
while True:
    print("""===============================
成绩分析系统
===============================
1.输入成绩数据
2.查看成绩统计
3.查看成绩排名
4.查看成绩分布
5.查询学生成绩
6.退出系统""")
    choice = input("请输入数字选择：")
    if choice == "1":
        input_score()
    elif choice == "2":
        show_statistics()
    elif choice == "3":
        show_rank()
    elif choice == "4":
        show_score_distribution()
    elif choice == "5":
        query_student()
    elif choice == "6":
        print("程序已退出，再见！")
        break
    else:
        print("输入无效，请选择1~6之间的数字！")
    input("\n按回车键返回菜单...")
```
